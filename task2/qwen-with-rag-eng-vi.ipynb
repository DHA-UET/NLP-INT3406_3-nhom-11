```python
import os
import torch
from unsloth import FastLanguageModel


DATA_DIR = "/kaggle/input/vlst-medical-mt-cleaned" 
TRAIN_EN_PATH = os.path.join(DATA_DIR, "train_ratio_filtered.en.txt")
TRAIN_VI_PATH = os.path.join(DATA_DIR, "train_ratio_filtered.vi.txt")
TEST_EN_PATH = os.path.join(DATA_DIR, "clean_public_test.en.txt")
TEST_VI_PATH = os.path.join(DATA_DIR, "clean_public_test.vi.txt")

MAX_SAMPLES = None
MAX_TEST_SAMPLES = None

MODEL_NAME = "unsloth/Qwen3-0.6B-unsloth-bnb-4bit"
MAX_SEQ_LENGTH = 4096 
TRAIN_MAX_LEN = 512
SEED = 3407

CHROMA_DB_DIR = "/kaggle/working/chroma_medical_en_vi"
N_RETRIEVAL = 3 

torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

print("="*60)
print("CONFIGURATION")
print("="*60)
print(f"Model: {MODEL_NAME}")
print(f"Max Seq Length: {MAX_SEQ_LENGTH}")
print(f"Train Max Length: {TRAIN_MAX_LEN}")
print(f"RAG Examples: {N_RETRIEVAL}")
print(f"CUDA: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
print("="*60)

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=MODEL_NAME,
    max_seq_length=MAX_SEQ_LENGTH,
    load_in_4bit=True,
    load_in_8bit=False,
    full_finetuning=False,
)

print("\nâœ“ Model loaded successfully!")
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=32,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj",
                    "gate_proj", "up_proj", "down_proj"],
    lora_alpha=32,
    lora_dropout=0.0,
    bias="none",
    use_gradient_checkpointing="unsloth",
    random_state=SEED,
    use_rslora=False,
    loftq_config=None,
)

trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"âœ“ LoRA adapters added")
print(f"âœ“ Trainable parameters: {trainable:,}")
print(f"âœ“ Model on device: {next(model.parameters()).device}")
```

```python
def load_parallel_data(en_path, vi_path, max_samples=None):
    """Load English-Vietnamese parallel data"""
    with open(en_path, "r", encoding="utf-8") as f:
        en_lines = [line.strip() for line in f]
    
    with open(vi_path, "r", encoding="utf-8") as f:
        vi_lines = [line.strip() for line in f]
    
    assert len(en_lines) == len(vi_lines), "Mismatch: EN and VI lines!"
    
    if max_samples:
        en_lines = en_lines[:max_samples]
        vi_lines = vi_lines[:max_samples]
    
    data = [(en, vi) for en, vi in zip(en_lines, vi_lines) if en and vi]
    
    print(f"Loaded {len(data):,} parallel pairs")
    return data

print("Loading training data...")
train_pairs = load_parallel_data(TRAIN_EN_PATH, TRAIN_VI_PATH, MAX_SAMPLES)

print("Loading test data...")
test_pairs = load_parallel_data(TEST_EN_PATH, TEST_VI_PATH, MAX_TEST_SAMPLES)

print(f"\nSample:")
print(f"  EN: {train_pairs[0][0]}")
print(f"  VI: {train_pairs[0][1]}")
```

```python
import chromadb
from sentence_transformers import SentenceTransformer
from tqdm.auto import tqdm
import torch


print("Loading embedding model...")
embed_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
embed_device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
embed_model = embed_model.to(embed_device)
print(f"âœ“ Embedding model on: {embed_device}")

chroma_client = chromadb.PersistentClient(path=CHROMA_DB_DIR)

collection_name = "medical_en_vi_pairs"
try:
    collection = chroma_client.get_collection(collection_name)
    print(f"âœ“ Loaded existing ChromaDB: {collection.count()} docs")
except:
    print("Building ChromaDB index with GPU embeddings...")
    
    collection = chroma_client.create_collection(
        name=collection_name,
        metadata={"hnsw:space": "cosine"}
    )
    
    all_en_texts = [pair[0] for pair in train_pairs]
    all_vi_texts = [pair[1] for pair in train_pairs]
    
    print(f"Computing {len(all_en_texts):,} embeddings on GPU...")
    embed_batch_size = 256 
    all_embeddings = []
    
    for i in tqdm(range(0, len(all_en_texts), embed_batch_size), desc="GPU Embedding"):
        batch_texts = all_en_texts[i:i+embed_batch_size]
        batch_emb = embed_model.encode(
            batch_texts,
            convert_to_numpy=True,
            show_progress_bar=False,
            batch_size=embed_batch_size,
            device=embed_device
        )
        all_embeddings.extend(batch_emb.tolist())

    print(f"âœ“ Computed {len(all_embeddings):,} embeddings")
    
    print("Adding to ChromaDB...")
    chroma_batch = 5000 
    
    for i in tqdm(range(0, len(train_pairs), chroma_batch), desc="Indexing"):
        end_idx = min(i + chroma_batch, len(train_pairs))
        
        collection.add(
            documents=all_en_texts[i:end_idx],
            embeddings=all_embeddings[i:end_idx],
            metadatas=[{"vi": vi} for vi in all_vi_texts[i:end_idx]],
            ids=[f"pair_{j}" for j in range(i, end_idx)]
        )
    
    print(f"âœ“ Indexed {collection.count():,} EN-VI pairs")
    print(f"âœ“ ChromaDB saved to: {CHROMA_DB_DIR}")
    
    test_query = train_pairs[100][0]
    print(f"\nTest query: {test_query[:80]}...")
    test_emb = embed_model.encode([test_query], device=embed_device)[0].tolist()
    results = collection.query(query_embeddings=[test_emb], n_results=3)
    print(f"Retrieved {len(results['documents'][0])} examples")
    
    print(f"\nâœ“ RAG system ready with GPU embeddings!")
```

```python
from datasets import Dataset

SYSTEM_PROMPT = "You are an expert medical translator from English to Vietnamese."

def format_training_sample(en_text, vi_text):
    """Format EN-VI pair for training with Qwen3 chat template"""
    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": f"Translate to Vietnamese: {en_text}"},
        {"role": "assistant", "content": vi_text}
    ]
    
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=False,
        enable_thinking=False,
    )
    return text

print("Preparing training data with filtering...")
train_texts = []
skipped = 0

for en, vi in tqdm(train_pairs, desc="Formatting"):
    text = format_training_sample(en, vi)
    token_count = len(tokenizer.encode(text, add_special_tokens=True))
    
    if token_count <= TRAIN_MAX_LEN:
        train_texts.append(text)
    else:
        skipped += 1

print(f"\nKept: {len(train_texts):,} samples")
print(f"Skipped: {skipped} samples (>{TRAIN_MAX_LEN} tokens)")

train_dataset = Dataset.from_dict({"text": train_texts})
train_dataset = train_dataset.shuffle(seed=SEED)

print(f"\nSample formatted text:")
print(train_dataset[0]["text"][:])
```

```python
from trl import SFTTrainer, SFTConfig

# === HYPERPARAMETERS
NUM_TRAIN_EPOCHS = 1
BATCH_SIZE = 32
GRAD_ACCUM = 4 
MAX_STEPS = 2000
LEARNING_RATE = 1e-4  
WARMUP_RATIO = 0.1  
LOGGING_STEPS = 10
SAVE_STEPS = 500

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=train_dataset,
    args=SFTConfig(
        dataset_text_field="text",
        max_seq_length=TRAIN_MAX_LEN,
        num_train_epochs = NUM_TRAIN_EPOCHS,
        
        # Batch & Gradient
        per_device_train_batch_size=BATCH_SIZE,
        gradient_accumulation_steps=GRAD_ACCUM,
        
        # Training schedule
        # max_steps=MAX_STEPS,
        warmup_ratio=WARMUP_RATIO,
        
        # Optimizer
        learning_rate=LEARNING_RATE,
        lr_scheduler_type="linear",
        optim="adamw_8bit",
        weight_decay=0.01,
        
        # Logging & Saving
        logging_steps=LOGGING_STEPS,
        save_strategy="steps",
        save_steps=SAVE_STEPS,
        save_total_limit=2,
        
        # Mixed precision
        fp16=True,
        
        seed=SEED,
        output_dir="./qwen3-0.6b-en-vi-medical",
        report_to="none",
    ),
)

print(f"\nTraining config:")
print(f"  Effective batch: {BATCH_SIZE * GRAD_ACCUM}")
# print(f"  Total steps: {MAX_STEPS}")
# print(f"  Warmup steps: {int(MAX_STEPS * WARMUP_RATIO)}")
# print(f"  Total samples: ~{MAX_STEPS *BATCH_SIZE * GRAD_ACCUM:,}")
print("\n" + "="*60)
print("STARTING TRAINING")
print("="*60)

trainer_stats = trainer.train()

print("\n" + "="*60)
print("TRAINING COMPLETED")
print("="*60)
print(f"Global steps: {trainer.state.global_step}")
print(f"Final loss: {trainer_stats.training_loss:.4f}")
```

```python
OUTPUT_DIR = "qwen3-0.6b-en-vi-medical-final"
os.makedirs(OUTPUT_DIR, exist_ok=True)

model.save_pretrained(os.path.join(OUTPUT_DIR, "lora_adapter"))
tokenizer.save_pretrained(os.path.join(OUTPUT_DIR, "tokenizer"))

print(f"âœ“ Model saved to: {OUTPUT_DIR}")
```

```python
from unsloth import FastLanguageModel

FastLanguageModel.for_inference(model)

def translate_en_to_vi_with_rag(en_text, n_examples=3, max_new_tokens=512):
    """
    Translate English to Vietnamese with RAG few-shot prompting
    
    Args:
        en_text: English input
        n_examples: Number of retrieved examples
        max_new_tokens: Max output length
    """
    query_emb = embed_model.encode(
        [en_text],
        convert_to_numpy=True,
        device=embed_device,
        show_progress_bar=False
    )[0].tolist()
    
    results = collection.query(
        query_embeddings=[query_emb],
        n_results=n_examples
    )
    
    few_shot_examples = ""
    for i, (en_ex, meta) in enumerate(zip(results['documents'][0], results['metadatas'][0])):
        vi_ex = meta['vi']
        few_shot_examples += f"""Example {i+1}:
English: {en_ex}
Vietnamese: {vi_ex}

"""
    
    user_content = f"""Here are some translation examples:

{few_shot_examples}Now translate this:
English: {en_text}
Vietnamese:"""
    
    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": user_content}
    ]
    
    prompt = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True,
        enable_thinking=False,
    )
    
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=MAX_SEQ_LENGTH)
    inputs = {k: v.to(model.device) for k, v in inputs.items()}
    
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id,
        )
    
    generated_ids = outputs[0][inputs['input_ids'].shape[1]:]
    translation = tokenizer.decode(generated_ids, skip_special_tokens=True).strip()
    
    return translation

print("Testing RAG-enhanced translation:\n")
for i in range(3):
    print(f"--- Example {i+1} ---")
    en_src = test_pairs[i][0]
    vi_ref = test_pairs[i][1]
    print(f"EN:   {en_src}")
    vi_pred = translate_en_to_vi_with_rag(en_src, n_examples=N_RETRIEVAL)
    print(f"PRED: {vi_pred}")
    print(f"REF:  {vi_ref}")
    print()
```

```python
import evaluate
from tqdm.auto import tqdm

print(f"Translating {len(test_pairs):,} test samples with RAG...\n")

predictions = []
references = []

for en_text, vi_ref in tqdm(test_pairs, desc="Translating"):
    vi_pred = translate_en_to_vi_with_rag(en_text, n_examples=N_RETRIEVAL)
    predictions.append(vi_pred)
    references.append(vi_ref)

metric = evaluate.load("sacrebleu")
bleu_result = metric.compute(
    predictions=predictions,
    references=[[ref] for ref in references]
)

print("\n" + "="*60)
print("EVALUATION RESULTS")
print("="*60)
print(f"BLEU Score: {bleu_result['score']:.2f}")
print(f"BLEU-1: {bleu_result['precisions'][0]:.2f}")
print(f"BLEU-2: {bleu_result['precisions'][1]:.2f}")
print(f"BLEU-3: {bleu_result['precisions'][2]:.2f}")
print(f"BLEU-4: {bleu_result['precisions'][3]:.2f}")
print("="*60)

print("\nSample predictions:\n")
for i in range(min(5, len(test_pairs))):
    print(f"Example {i+1}:")
    print(f"  EN:   {test_pairs[i][0][:80]}...")
    print(f"  REF:  {references[i][:80]}...")
    print(f"  PRED: {predictions[i][:80]}...")
    print()
```
